```python
# import the libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

```python
# get the image and assign a varibale to it
pic = Image.open('../DATA/00-puppy.jpg')
pic
```

```python
# to see the type of the variable
type(pic)
```

```python
# create an array of the pic
pic_arr = np.asarray(pic)
pic_arr.shape
```

```python
# use the matplotlib to show the image using arrays
plt.imshow(pic_arr)
```

```python
# copying one array to another
pic1 = pic_arr.copy()
pic1
```

```python
plt.imshow(pic1)
```

```python
# making second column into zero to change the colour
pic1[:,:,1] = 0
plt.imshow(pic1)
```

```python
# changing the third column as zero
pic2 = pic1.copy()
pic2[:,:,2] = 0
plt.imshow(pic1)
```

```python
plt.imshow(pic2)
```
