```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/00-puppy.jpg')
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
plt.imshow(img)
```

```python
img2 = cv2.imread('../DATA/watermark_no_copy.png')
plt.imshow(img2)
```

```python
img = cv2.resize(img, (1200, 1200))
img2 = cv2.resize(img2, (1200, 1200))
```

```python
# we want to blend both images
blended_img = cv2.addWeighted(src1 = img, alpha = 0.7, src2 = img2, beta = 0.3, gamma= 0)
plt.imshow(blended_img)
```

```python
def load_img():
    img = cv2.imread('../DATA/bricks.jpg').astype(np.float32) / 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
display_img(i)
```

```python
def apply_text(img):
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img,text = "ABCD", org = (50,300), fontFace = font, fontScale=5, color=(255,0,0), thickness=5)
    return img
```

```python
apply_text(img)
plt.imshow(img)
```
