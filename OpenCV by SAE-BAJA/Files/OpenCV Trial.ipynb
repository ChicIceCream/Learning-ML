```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/00-puppy.jpg')
# In opencv, the format of an image is BGR instead of RGB
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
img_gray.shape
```

```python
img_resized = cv2.resize(img_gray, (1000, 800))
plt.imshow(img_resized)
```

```python
# now we will flip the image
flipped_img = cv2.flip(img_rgb, 1)
plt.imshow(flipped_img)
```

```python
cv2.imwrite('flipped.jpg', flipped_img)
```

```python
fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111)
ax.imshow(flipped_img)
```
